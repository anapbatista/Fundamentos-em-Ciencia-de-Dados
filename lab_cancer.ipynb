```python
import pandas as pd
from sklearn.datasets import load_breast_cancer

data = load_breast_cancer()
df = pd.DataFrame(data.data, columns=data.feature_names)

X = df.values
y = data.target

#estatísticas descritivas básicas
print(df.columns)
print(df.describe())
print(df.info())

#tamanho da matriz valores numericos e tamanho da matriz target
print(X.shape)
print(y.shape)
df
```

```python
from sklearn.model_selection import StratifiedKFold
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, f1_score
import numpy as np

folds = StratifiedKFold(n_splits=10, shuffle=True, random_state=20)
fold_num=0

# Definindo os valores de k
k_values = np.arange(1, 16, 2)  # k de 1 a 15, pulando de dois em dois

# Dicionário para armazenar os resultados
results = {}
cont = 0

# Iterando sobre cada fold
for train_index, val_index in folds.split(X, y):
    fold_num+=1
    X_train_fold, X_val_fold = X[train_index], X[val_index]
    y_train_fold, y_val_fold = y[train_index], y[val_index]

    # Iterando sobre cada valor de k
    for k in k_values:

        cont+=1
        # Lista para armazenar as métricas de cada fold
        accuracy_scores = []
        f1_scores = []

        # Instanciando o classificador k-NN com o valor atual de k
        knn_classifier = KNeighborsClassifier(n_neighbors=k)
        knn_classifier.set_params(weights="uniform").fit(X_train_fold, y_train_fold)

        # Realizando previsões no conjunto de validação
        y_pred = knn_classifier.predict(X_val_fold)

        # Calculando a acurácia e F1
        accuracy = accuracy_score(y_val_fold, y_pred)
        f1 = f1_score(y_val_fold, y_pred)


        # Armazenando os resultados
        accuracy_scores.append(accuracy)
        f1_scores.append(f1)


        # Armazenando as métricas no dicionário
        results[cont] = {'fold': fold_num, 'k': k, 'accuracy': accuracy, 'f1': f1}


# Convertendo resultados em DataFrame
results_df = pd.DataFrame.from_dict(results, orient='index')

#com base nesses resultados, é possível ter uma noção de erro para cada K
results_df
```
