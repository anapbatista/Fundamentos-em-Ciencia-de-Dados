trabalhar com dataset california

```python
df=pd.read_csv('sample_data/california_housing_train.csv')
df.columns
df.describe()
df.corr()
df.cov()
```

```python
df.plot.scatter(x='median_income', y='median_house_value')
df.plot.scatter(x='longitude', y='latitude')
```

```python
df=df.sort_values(['median_house_value'])
df
```

```python
df_01=df.copy()
max=df_01.max()
min=df_01.min()
df_01=(df_01 - min)/(max-min)
df_01.describe()
```

```python
df_zscore=df.copy()
media=df.mean()
desvio=df.std()
df_zscore=(df_zscore-media)/desvio
df_zscore
df_zscore.describe()
```

```python
from scipy.stats import zscore
df3=zscore(df)
df3.describe()
```

```python
from scipy.spatial.distance import pdist, squareform
dist1=pdist(df)
plt.imshow(squareform(dist1))
```

```python
dist2=pdist(df_01)
plt.imshow(squareform(dist2))
```

```python
dist3=pdist(df3)
plt.imshow(squareform(dist3))
```
